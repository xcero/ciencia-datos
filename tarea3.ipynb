```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


df = pd.read_csv('Titanic1.csv', header=0)
#/


while True: #Generamos el Ciclo repetitivo para el menu
    
    os.system("cls") #Limpieza de pantalla
    print("Menu datos sobre el Titanic:")
    print("1. Ver graficas de sobrevivientes y Muertos")
    print("2. Ver precios de Ticket max y min")
    print("3. Buscar Registro por Ticket (11769 ,PC 17609)")
    print("4. Salir")

    opcion = input("Ingrese su opción: ") #Capturamos Opción
    
    if opcion == "1":
        Muertos= df[df['Sobrevivio'] == 0]
        Vivos= df[df['Sobrevivio']==1]

        print("Muertos en el Titanic:", len(Muertos.index))
        print("Sobrevivientes en el Titanic:", len(Vivos.index))


        y = np.array([len(Muertos.index), len(Vivos.index)])

        mylabels = ["Muertos", "Vivos"]
        plt.pie(y, labels = mylabels, shadow = True)

        plt.legend(title = "Titanic:")
        plt.show()

        MuertosxClase=Muertos.groupby('Clase')['Sobrevivio'].count()
        VivosxClase=Vivos.groupby('Clase')['Sobrevivio'].count()

        print("Primera clase-- Muertos:", MuertosxClase[1], "-- Vivos", VivosxClase[1])
        print("Segunda clase-- Muertos:", MuertosxClase[2], "-- Vivos", VivosxClase[2])
        print("Tercera clase-- Muertos:", MuertosxClase[3], "-- Vivos", VivosxClase[3])
        print("-------------------")
        print()

        MuertosxClase= MuertosxClase.plot(kind='bar')
        plt.show()

        VivosxClase= VivosxClase.plot(kind='bar')
        plt.legend(title = "Vivos por Clase:")
        plt.show()
        input()
        continue

    elif opcion == "2":
        Tarifa =  df[df['Tarifa'].notnull() & df['Tarifa'] > 0]
        print("Tickets vendidos", len(Tarifa.index))

        print(Tarifa.min())
        print(Tarifa.max())
        input()
        continue
    elif opcion == "3":
        Numero= input("Digite el numero de ticket a buscar: ")
        Busqueda =df[df['Ticket'] == Numero]
        print(Busqueda)
        input()
        continue

    elif opcion == "4":
        print("Saliendo del Sistema")
        input()
        break  #saliendo del ciclo repetitivo

    else:
        print("Opción incorrecta")
        input()
        continue     #Continuando el ciclo repetitivo
```
